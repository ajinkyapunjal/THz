```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
filename = 'phasecorrection.txt'
```

```python
df = pd.read_csv(filename,sep='\s+')
```

```python
df
```

```python
a = df.to_numpy()

for ii in range(a.shape[1]):
    for i in range(a.shape[0]-1):
        diff = abs(a[i,ii])-abs(a[i+1,ii])
        if diff>175 and abs(a[i,ii])>abs(a[i+1,ii]):
            a[i+1:,ii] = a[i+1:,ii]-360   
```

```python
plt.plot(a)
plt.show()
```

```python
b = pd.DataFrame(a)
b.to_csv('raw_data.csv', index=False)
```
